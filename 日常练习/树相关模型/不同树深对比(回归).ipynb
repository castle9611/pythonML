```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'
```

```python
# 随机数种子
rng = np.random.RandomState(1) 
# 生成0~5之间随机的x的取值
X = np.sort(5 * rng.rand(80,1), axis=0) 
# y = np.sin(X).ravel() #生成正弦曲线(注意要扁平化,这里用flatten()和reval()都可以)
y = np.sin(X).flatten()
y[::5] += 3 * (0.5 - rng.rand(16)) #每隔五个位置，在正弦曲线上加噪声
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",c="darkorange", label="data")
plt.show()
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

```python
# X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]# 效果等同于reshape（），如下：
X_test = np.arange(0.0, 5.0, 0.01).reshape(-1,1)
```

np.arange(),生成的是（n,）大小的数据，需要用的是（n,1）大小，所以要转换一下。

```python
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y, s=20, edgecolor="black",c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",label="max_depth=2", lw=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", lw=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
